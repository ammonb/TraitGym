```python
import pandas as pd
```

```python
dataset = "mendelian_traits_matched_9"
base_dir = f"../../results/dataset/{dataset}"
V = pd.read_parquet(f"{base_dir}/test.parquet")
V
```

```python
models = [
    "GPN-MSA_LLR.minus.score",
    "Enformer_L2_L2.plus.all",
    "Borzoi_L2_L2.plus.all",
    "CADD.plus.RawScore",

    "GPN_LLR.minus.score",
    "NucleotideTransformer_LLR.minus.score",
]

for model in models:
    df = (
        pd.read_parquet(f"{base_dir}/preds/all/{model}.parquet")
        .rename(columns={"score": model})
    )
    V = pd.concat([V, df], axis=1)
V
```

```python
variant_pos = 113105781
match_group = V[V.pos==variant_pos].match_group.iloc[0]
V = V[V.match_group==match_group]
V
```

```python
for model in models:
    V[f"rank_{model}"] = V[model].rank(ascending=False)
V.loc[V.label, [f"rank_{model}" for model in models]]
```
