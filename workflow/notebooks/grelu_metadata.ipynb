```python
import grelu.resources
import pandas as pd
```

```python
project, model_name = "enformer", "human"
metadata = pd.DataFrame(
    grelu.resources.load_model(
        project=project, model_name=model_name
    ).data_params['tasks']
)
metadata.to_csv("Enformer_metadata.csv", index=False)
metadata
```
