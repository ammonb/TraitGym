```python
import bioframe as bf
import pandas as pd
```

```python
#dataset = "mendelian_traits_matched_9"
dataset = "complex_traits_matched_9"
V = pd.read_parquet(f"../../results/dataset/{dataset}/test.parquet")
V
```

```python
tss = pd.read_parquet("../../results/tss.parquet")
tss
```

```python
V["start"] = V.pos - 1
V["end"] = V.pos
V = bf.closest(V, tss).rename(columns={
    "gene_id_": "gene",
}).drop(columns=["start", "end", "chrom_", "start_", "end_", "distance"])
V
```

```python
#V.to_csv("mendelian_traits.csv.gz", index=False)
V.to_csv("complex_traits.csv.gz", index=False)
```
